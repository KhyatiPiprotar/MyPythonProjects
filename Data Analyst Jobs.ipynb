```python
# importing different libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Loading Data analyst jobs in USA dataset
df = pd.read_excel("DataAnalystFinalP_excel.xlsx",sheet_name = "in")
# Show head of the data
df.head()
```

```python
#Show column names
df.info()
```

```python
#Findout null values
df.isnull().sum() #There is no null values because all the null values are replaced by 0 in this dataset
```

```python
# Howmany rows and columns are in the Dataset? 
df.shape
```

```python
# Find out number of industries that pays $113K-$132K and list Industry names with counts.
# We can say that Insurance Carriers industry has the maximum jobs available with very good pay range. 

df[df['Salary Estimate(Glassdoor)']== '$113K-$132K'].groupby('Industry').count()['Location']
```

```python
#Find out the state that has maximum number of job postings.
#As you can see, California has 625 jobs available.

JobsStates = df.groupby('State').count().sort_values(by = 'Job Title',ascending = False)
JobsStates
```

```python

plt.figure(figsize = (10,8))
plt.xlabel('States')
plt.title('States Vs Avg. salaries')
palette = np.repeat(np.array(sns.color_palette("mako")),3,axis=0)
sns.barplot('State','Avg*1000', data = df, palette = palette)
```

```python
#Remove the rows where ratings are missing( are 0)
ZeroRatings = df[~(df['Rating']== 0)]
ZeroRatings
```

```python
#Findout companies with 4.5 and 5 stars ratings
#Staffigo Technical Services, LLC\n5.0 has 58 ratings, with 4.5 or 5 stars
ZeroRatings[ZeroRatings['Rating'].isin([4.5,5])].groupby('Company Name').count().sort_values(by = 'Rating',ascending = False)
```

```python
df.head()
```

```python
#Separate all the city names in the column and find out howmany jobs available in Los Angeles
df['City'] = df['Location'].apply(lambda x : x.split(',')[0])
df[df['City']== 'Los Angeles'].count()
```

```python
#Find correlation between Company Ratings and Average Pay
#There is no relationship between them as you can see

col1 = df['Rating']
col2 = df['Avg*1000']
correlation = col1.corr(col2)
print(correlation)
```

```python
#Will plot a scatter plot to visualize the correlation between Rating and Avg. salary

plt.figure(figsize= (12,10))
sns.scatterplot(x = 'Rating', y= 'Avg*1000', data =df)
```

```python
#Find correlation between all the variables

df.corr(method ='pearson')
```

```python
#Plot correlation matrix
plt.figure(figsize = (12,10))
plt.title("Correlation matric for numeric variables")
corr_matrix = df.corr(method ='pearson')
sns.heatmap(data = corr_matrix, annot = True)
```

```python
#Which company has maximum jobs available?

df.groupby(by = 'Company Name').count().sort_values(by = 'State',ascending = False)['Job Title']
```

```python
#Find out maximum salary range. Before we do that, we need to remove '0'(which is string) and 0 values

df3 = df[~(df['Salary Estimate(Glassdoor)'].isin(['0',"'0'"]))]
df3
```

```python
#After removing those. We can find max salary range with simple max() function.

df3['Salary Estimate(Glassdoor)'].max()
```
